```python
%matplotlib inline
import requests
from bs4 import BeautifulSoup
import os
import pandas as pd
import time
import re
import seaborn
import matplotlib
import time
from random import choice 
from time import sleep
from random import uniform 

%config InlineBackend.figure_format = 'svg'
from pylab import rcParams
rcParams['figure.figsize'] = 8, 5\


s = requests.Session() 
s.headers.update({
        'Referer': 'https://github.com',
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.9; rv:45.0) Gecko/20100101 Firefox/45.0'
    })
total_pages = 0 
```

```python
def load_user_data(page, useragent = None,  proxy = None):
    url = 'https://github.com/search?p=%d&q=dataset+music&type=Repositories'% (page)
    request = requests.get(url, headers = useragent, proxies = proxy)
    return request.text

def contain_datasets_data(text):
    soup = BeautifulSoup(text)
    dataset_list = soup.find('ul', {'class': 'repo-list'})
    return dataset_list is not None

def read_file(filename):
    with open(filename) as input_file:
        text = input_file.read()
    return text

def total_pages(text):
    soup = BeautifulSoup(text)
    pages = soup.find('em').get('data-total-pages')
    return pages is not None

def parse_user_datafile_bs(filename):
    results = []
    text = read_file(filename)   
    soup = BeautifulSoup(text)
    dataset_list = soup.find('ul', {'class': 'repo-list'})
    items = dataset_list.find_all('li', {'class': 'repo-list-item'})
    for item in items:
        # getting movie_id
        dataset_link = item.find('h3').find('a').get('href')
        dataset_name = item.find('h3').find('a').text
        dataset_link = 'https://github.com' + dataset_link
        
        soup = BeautifulSoup(text)
        text_about = soup.find('div', {'class': 'Box-body'}).text
       
            
        results.append({
                'dataset_link': dataset_link,
                'dataset_name': dataset_name
            })
    return results
```

```python
def checkIP():
    ip = requests.get('http://checkip.dyndns.org').content
    soup = BeautifulSoup(ip, 'html.parser')
    print(soup.find('body').text)
checkIP()
```

```python
useragents = open('useragents.txt').read().split('\n')
proxies = open('proxy-list.txt').read().split('\n')
```

```python
page = 1

while True:
    proxy = {'http': 'http://' + choice(proxies) }
    useragent = {'User-Agent': choice(useragents) }
    print(proxy)
    print(useragent)
    print ('Loading page #%d' % (page))
    data = load_user_data(page, useragent, proxy)
    a = uniform(3,6)
    print(a)
    sleep(a)
    if contain_datasets_data(data):
        with open('./datasets_video/page_%d.html' % (page), 'wb') as output_file:
            
            output_file.write(data.encode('utf-8'))
            page += 1
    else:
        continue
        
            
```

```python
results = []
for filename in os.listdir('./datasets_video/'):
    results.extend(parse_user_datafile_bs('./datasets_video/' + filename))
    print(filename)
```

```python
user_data_df = pd.DataFrame(results)
```
