```python
import pandas as pd
import csv
%matplotlib inline
import requests
from bs4 import BeautifulSoup
import os
import time
import re
import seaborn
import matplotlib
import time
from random import choice 
from time import sleep
from random import uniform 

```

```python
links_of_datasets = pd.read_csv("dropbox_datasets_raw_video" ,sep=' ', encoding='utf-8')
```

```python
def load_user_data(url, useragent = None,  proxy = None):
    request = requests.get(url, headers = useragent, proxies = proxy)
    return request.text

def contain_datasets_data(text):
    soup = BeautifulSoup(text)
    text_about = soup.find('div', {'class': 'Box-body'})
    return text_about is not None

def read_file(filename):
    with open(filename) as input_file:
        text = input_file.read()
    return text

def parse_user_datafile_bs(filename):
    results = []
    text = read_file(filename)   
    soup = BeautifulSoup(text)
    text_about = soup.find('div', {'class': 'Box-body'}).text
    dataset_name = soup.find('div', {'class': 'Box-body'}).find('h1').text
    print(dataset_name)
    results.append({
            'dataset_name': dataset_name,
             'text': text_about
            })
    return results
```

```python
useragents = open('useragents.txt').read().split('\n')
proxies = open('proxy-list.txt').read().split('\n')
```

```python
num = 1
for index, row in links_of_datasets.iterrows():
    if index >= num-1:
        proxy = {'http': 'http://' + choice(proxies) }
        useragent = {'User-Agent': choice(useragents) }
        print(proxy)
        print(useragent)
        data = load_user_data(row["dataset_link"], useragent, proxy)
        #a = uniform(3,6)
        #print(a)
        print(row["dataset_name"])
        print(index)
        #sleep(a)
        if contain_datasets_data(data):
            with open('./dataset_info_video/dataset%d.html' % (num), 'wb') as output_file:
                output_file.write(data.encode('utf-8'))
                num += 1
        else:
            continue
    else:
        print(index)
   
```

```python
results = []
for filename in os.listdir('./dataset_info_video/'):
    try:
        results.extend(parse_user_datafile_bs('./dataset_info_video/' + filename))
        print(filename)
    except:
        print('miss')
        continue
    
```

```python
datasets_info = pd.DataFrame(results)
```

```python
datasets_info.to_csv('dropbox_datasets_video', sep=' ', encoding='utf-8')
```
