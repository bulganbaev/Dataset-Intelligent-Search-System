```python
from NaiveBayes import  Pool
import os

DClasses = ["text",  "image",  "video"]

base = "descriptions_dataset/"
p = Pool()
for i in DClasses:
    p.learn(base + i, i)
```

```python
base = "descriptions/"
dir = os.listdir(base)
for file in dir:
    res = p.Probability(base + "/" + file)
    print( ": " + file + ": " + str(res))
```
