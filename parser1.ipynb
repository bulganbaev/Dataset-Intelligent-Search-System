```python
%matplotlib inline
import requests
from bs4 import BeautifulSoup
import os
import pandas as pd
import time
import re
import seaborn
import matplotlib
import time
from random import choice 
from time import sleep
from random import uniform 

num = 1 #global variable
```

```python
def load_user_data(page, useragent = None,  proxy = None):
    url = 'https://github.com/search?p=%d&q=dataset+faces&type=Repositories'% (page)
    request = requests.get(url, headers = useragent, proxies = proxy)
    return request.text

def contain_datasets_data(text):
    soup = BeautifulSoup(text)
    dataset_list = soup.find('ul', {'class': 'repo-list'})
    return dataset_list is not None

def read_file(filename):
    with open(filename) as input_file:
        text = input_file.read()
    return text

def parse_user_datafile_bs(data, useragent=None, proxy=None):
    results = [] 
    global num
    soup = BeautifulSoup(data)
    dataset_list = soup.find('ul', {'class': 'repo-list'})
    items = dataset_list.find_all('li', {'class': 'repo-list-item'})
    for item in items:
        # getting movie_id
        dataset_link = item.find('h3').find('a').get('href')
        dataset_name = item.find('h3').find('a').text
        dataset_link = 'https://github.com' + dataset_link
        
        request_temp = requests.get(dataset_link, headers = useragent, proxies = proxy)
        text_temp = request_temp.text
        soup_temp = BeautifulSoup(text_temp)
        try:    
            text_about = soup_temp.find('div', {'class': 'Box-body'}).text
            document = re.sub(r'\W', ' ', str(text_about))
            document = re.sub(r'\s+[a-zA-Z]\s+', ' ', document)
            document = re.sub(r'\^[a-zA-Z]\s+', ' ', document)
            document = re.sub(r'\s+', ' ', document, flags=re.I)
            document = re.sub(r'^b\s+', '', document)
            document = document.lower()
            file = open(r"./data/type/faces/%d.txt" %num,"w+") 
            file.write(document)
            num += 1
            file.close()
        except:
            print(text_about)
            continue
    
    return results
```

```python
def checkIP():
    ip = requests.get('http://checkip.dyndns.org').content
    soup = BeautifulSoup(ip, 'html.parser')
    print(soup.find('body').text)
checkIP()
```

```python
useragents = open('./data/headers/useragents.txt').read().split('\n')
proxies = open('./data/headers/proxy-list.txt').read().split('\n')
```

```python
page = 1

while True:
    proxy = {'http': 'http://' + choice(proxies) }
    useragent = {'User-Agent': choice(useragents) }
    print(proxy)
    print(useragent)
    print ('Loading page #%d' % (page))
    data = load_user_data(page, useragent, proxy)
    a = uniform(3,6)
    print(a)
    sleep(a)
    if contain_datasets_data(data):
        parse_user_datafile_bs(data, useragent, proxy)
        page += 1
    else:
        
        continue
        
            
```

```python
results = []
for filename in os.listdir('./datasets_video/'):
    results.extend(parse_user_datafile_bs('./datasets_video/' + filename))
    print(filename)
```

```python
user_data_df = pd.DataFrame(results)
```
